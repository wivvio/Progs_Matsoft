```python
df = pd.DataFrame([[max(arr_1), st.mean(mas_1), max(st.multimode(mas_1)), c1], 
                    [max(arr_2), st.mean(mas_2), max(st.multimode(mas_2)), c2],
                    [max(arr_3), st.mean(mas_3), max(st.multimode(mas_3)), c3],
                    [max(arr_4), st.mean(mas_4), max(st.multimode(mas_4)), c4],
                    [max(arr_5), st.mean(mas_5), max(st.multimode(mas_5)), c4],
                    [max(arr_6), st.mean(mas_6), max(st.multimode(mas_6)), c6]],
                   columns=['maxim', 'mean_norm', 'maxim_mode', 'peaks'])
df.to_excel("Clasters.xlsx")  
```

```python
masin = np.array([[max(arr_1), st.mean(mas_1), max(st.multimode(mas_1)), c1], 
                    [max(arr_2), st.mean(mas_2), max(st.multimode(mas_2)), c2],
                    [max(arr_3), st.mean(mas_3), max(st.multimode(mas_3)), c3],
                    [max(arr_4), st.mean(mas_4), max(st.multimode(mas_4)), c4],
                    [max(arr_5), st.mean(mas_5), max(st.multimode(mas_5)), c5],
                    [max(arr_6), st.mean(mas_6), max(st.multimode(mas_6)), c6]])
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=0).fit(masin)
df['cluster'] = kmeans.fit_predict(df[['maxim', 'mean_norm', 'maxim_mode', 'peaks']])
colors = ['#DF2020', '#81DF20']
df['c'] = df.cluster.map({0:colors[0], 1:colors[1]})
```

```python
plt.scatter(df.maxim, df.mean_norm, c = df.c, alpha = 0.9, s=50)
plt.title('Кластеризация больных\n', loc='left', fontsize=22)
plt.xlabel('maxim')
plt.ylabel('mean_norm')
print("\n")
```
